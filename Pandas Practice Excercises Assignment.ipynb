```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
filename = ['states0.csv', 'states1.csv', 'states2.csv', 'states3.csv', 'states4.csv', 
           'states5.csv', 'states6.csv', 'states7.csv', 'states8.csv', 'states9.csv']
```

```python
all_files = glob.glob("states*.csv")
```

```python
li = []
```

```python
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
```

```python
us_census = pd.concat(li, axis=0, ignore_index=True)
```

```python
us_census.head()
```

```python
us_census.columns
```

```python
 us_census.loc[:, 'Income'] = us_census.loc[:, 'Income'].str.replace('$', "")
```

```python
us_census.head(10)
```

```python
us_census[['Men', 'Women']] = us_census.GenderPop.str.split('_', expand=True)
```

```python
us_census.head(10)
```

```python
us_census.loc[:, 'Men'] = us_census.loc[:, 'Men'].str.replace("M", "")
us_census.loc[:, 'Women'] = us_census.loc[:, 'Women'].str.replace("F", "")
us_census.head(10)
```
